# PG-RAD compared to others

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
source_strengths = [100, 250, 500, 1000, 2500, 5000, 10000]
dist = [10,25,50,75,100]

# 50m from road
pg_rad_ss_max_cps = [3.46, 8.65, 17.29, 34.59, 86.47, 172.94, 345.88]
robert_ss_max_cps = [3.95, 9.87, 19.74, 39.48, 98.71, 197.4, 394.8]
robert_ss_max_cps_unc = [1.98, 3.14, 4.44, 6.28, 9.95, 14.05, 9.97]
alex_ss_max_cps = [3.31, 8.28, 16.56, 33.11, 82.79, 165.58, 331.1]

# 1000 MBq
pg_rad_dist_max_cps = [1266.09, 175.95, 34.59, 12.08, 5.34]
robert_dist_max_cps = [1451, 201.0, 39.48, 13.79, 6.1]
robert_dist_max_cps_unc = [38.7, 14.18, 6.28, 3.71, 2.5]
alex_dist_max_cps = [832.17, 152.99, 33.12, 11.86, 5.29]
```

```python
plt.figure(figsize=(12, 5))

# Plot 1: Source Strengths
plt.subplot(1, 2, 1)
plt.errorbar(source_strengths, robert_ss_max_cps, yerr=robert_ss_max_cps_unc, label='Excel program', marker='o', capsize=5)
plt.plot(source_strengths, pg_rad_ss_max_cps, label='PG-RAD', marker='o')
plt.plot(source_strengths, alex_ss_max_cps, label='ga-data-gen', marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Activity (MBq)')
plt.ylabel('Max CPS')
plt.title('Max CPS vs source activity')
plt.legend()
plt.grid(True, which="both", ls="--")

# Plot 2: Distances
plt.subplot(1, 2, 2)
plt.errorbar(dist, robert_dist_max_cps, yerr=robert_dist_max_cps_unc, label='Excel program', marker='o', capsize=5)
plt.plot(dist, pg_rad_dist_max_cps, label='PG-RAD', marker='o')
plt.plot(dist, alex_dist_max_cps, label='ga-data-gen', marker='o')
#plt.xscale('log')
plt.yscale('log')
plt.xlabel('Distance (m)')
plt.ylabel('Max CPS')
plt.title('Max CPS v.s. distance (A = 1000 MBq)')
plt.legend()
plt.grid(True, which="both", ls="--")

plt.tight_layout()
```
